```python
## Suppress warnings for cleaner outputs
import warnings
warnings.filterwarnings('ignore')


import numpy as np
import pandas as pd

import statsmodels.api as sm
import statsmodels.formula.api as smf

from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestRegressor
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline


# LOAD DATA

import pandas as pd
df = pd.read_csv('/content/eco.csv')
df.head()


# Convert Date to datetime and create time variables
df["Date"] = pd.to_datetime(df["Date"])
df["year"] = df["Date"].dt.year
df["month"] = df["Date"].dt.month

# Sort for safety
df = df.sort_values(["Country", "Date"]).reset_index(drop=True)

print("Data preview:")
print(df.head())
print("\nColumns:", df.columns.tolist())
```

```python
### Suppress warnings for cleaner outputs
import warnings
warnings.filterwarnings('ignore')


# DIFFERENCE-IN-DIFFERENCES (DiD) SETUP

# Compute country-level average interest rate
country_mean_ir = df.groupby("Country")["Interest Rate (%)"].mean()

# Threshold: 75th percentile of average interest rates
threshold = country_mean_ir.quantile(0.75)

treated_countries = country_mean_ir[country_mean_ir >= threshold].index.tolist()

print("\nTreated (high-interest) countries (top 25% by avg rate):")
print(treated_countries)

# Create treatment indicators
df["treated"] = df["Country"].isin(treated_countries).astype(int)

# Define post-policy period, e.g., from 2017 onward
df["post"] = (df["year"] >= 2017).astype(int)

# DiD interaction term
df["did"] = df["treated"] * df["post"]

# RUN DiD REGRESSION

# GDP Growth ~ DiD + controls + country FE + year FE

formula_did = (
    "Q('GDP Growth Rate (%)') ~ did "
    "+ Q('Inflation Rate (%)') "
    "+ Q('Unemployment Rate (%)') "
    "+ Q('Stock Index Value') "
    "+ C(Country) + C(year)"
)

did_model = smf.ols(formula_did, data=df)
did_results = did_model.fit(
    cov_type="cluster",
    cov_kwds={"groups": df["Country"]}  # cluster by country
)

print("\n DiD RESULTS")
print(did_results.summary())

did_coef = did_results.params.get("did", np.nan)
print(f"\nDiD estimate (did coefficient): {did_coef:.4f}")
```

The Difference-in-Differences estimate (did = 0.478) is positive but statistically insignificant (p = 0.472), indicating no evidence of a causal effect of high-interest-rate status on GDP growth. The confidence interval includes zero, and the model explains limited variation in GDP growth (R² = 0.044), which is typical for macro-panel growth regressions. Country fixed effects capture most persistent growth differences, while inflation, unemployment, and stock market controls are insignificant after conditioning on fixed effects. Cluster-robust standard errors are used, and the large condition number suggests multicollinearity, likely inflating standard errors but not biasing estimates.






```python

# DOUBLE MACHINE LEARNING (DML) FOR INTEREST RATE -> GDP GROWTH



# Outcome, treatment and features
Y = df["GDP Growth Rate (%)"].values
D = df["Interest Rate (%)"].values

feature_cols = [
    "Inflation Rate (%)",
    "Unemployment Rate (%)",
    "Stock Index Value",
    "year",
    "Country"
]
X = df[feature_cols].copy()

numeric_features = ["Inflation Rate (%)", "Unemployment Rate (%)", "Stock Index Value", "year"]
categorical_features = ["Country"]

preprocess = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numeric_features),
        ("cat", OneHotEncoder(drop="first"), categorical_features),
    ]
)

# Use Random Forest as the base learner for both Y|X and D|X
base_learner_y = RandomForestRegressor(
    n_estimators=200,
    max_depth=6,
    random_state=42,
    n_jobs=-1,
)

base_learner_d = RandomForestRegressor(
    n_estimators=200,
    max_depth=6,
    random_state=42,
    n_jobs=-1,
)

pipe_y = Pipeline(steps=[("preprocess", preprocess),
                        ("model", base_learner_y)])

pipe_d = Pipeline(steps=[("preprocess", preprocess),
                        ("model", base_learner_d)])
# Cross-fitting
kf = KFold(n_splits=3, shuffle=True, random_state=42)

Y_res = np.zeros(len(df))
D_res = np.zeros(len(df))

for fold, (train_idx, test_idx) in enumerate(kf.split(X), 1):
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    Y_train, Y_test = Y[train_idx], Y[test_idx]
    D_train, D_test = D[train_idx], D[test_idx]

    # Fit nuisance models on training folds
    pipe_y.fit(X_train, Y_train)
    pipe_d.fit(X_train, D_train)

    # Predict on held-out fold
    Y_hat = pipe_y.predict(X_test)
    D_hat = pipe_d.predict(X_test)

    # Store residuals
    Y_res[test_idx] = Y_test - Y_hat
    D_res[test_idx] = D_test - D_hat


# SECOND-STAGE OLS: RESIDUAL REGRESSION (DML EFFECT)


dml_df = pd.DataFrame({
    "Y_res": Y_res,
    "D_res": D_res
})

# Add constant term for OLS
dml_model = sm.OLS(dml_df["Y_res"], sm.add_constant(dml_df["D_res"]))
dml_results = dml_model.fit()

print("\nDML RESULTS ")
print(dml_results.summary())

theta_hat = dml_results.params["D_res"]
print(f"\nDML estimate of causal effect of Interest Rate on GDP Growth: {theta_hat:.4f}")
```

The Double Machine Learning (DML) estimate of the causal effect of interest rates on GDP growth is 0.068, and is not statistically significant (p = 0.242). The confidence interval includes zero, indicating no evidence of a causal relationship after flexibly controlling for confounders. The low R² (0.003) is expected in residual-on-residual regressions and does not reflect poor model performance. Overall, the results suggest that once nonlinear and high-dimensional controls are partialled out, interest rates do not have a statistically detectable impact on GDP growth.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Compute avg GDP growth by year for treated vs control
parallel_df = df.groupby(["year", "treated"])["GDP Growth Rate (%)"].mean().reset_index()
parallel_df["treated_label"] = parallel_df["treated"].map({0:"Control", 1:"Treated"})

plt.figure(figsize=(10,6))
sns.lineplot(data=parallel_df, x="year", y="GDP Growth Rate (%)", hue="treated_label", marker="o")
plt.axvline(2017, color="red", linestyle="--", label="Policy Start (2017)")
plt.title("Parallel Trends: Treated vs Control GDP Growth Over Time")
plt.ylabel("Mean GDP Growth Rate (%)")
plt.xlabel("Year")
plt.legend()
plt.grid(True)
plt.show()
```

**Interpretation:**

This figure plots average GDP growth rates for treated (high-interest) and control countries over time. The vertical dashed line marks the policy intervention in 2017. Prior to the intervention (2010–2016), the treated and control groups exhibit broadly similar trends in GDP growth, supporting the parallel trends assumption underlying the Difference-in-Differences design. Post-2017, the series display some divergence, though subsequent regression results indicate that these differences are not statistically significant.

```python
before_after = df.groupby(["post", "treated"])["GDP Growth Rate (%)"].mean().reset_index()
before_after["Period"] = before_after["post"].map({0:"Pre-Policy", 1:"Post-Policy"})
before_after["Group"] = before_after["treated"].map({0:"Control", 1:"Treated"})

plt.figure(figsize=(8,6))
sns.barplot(data=before_after, x="Period", y="GDP Growth Rate (%)", hue="Group")
plt.title("GDP Growth Before vs After Policy (Treated vs Control)")
plt.ylabel("Mean GDP Growth (%)")
plt.grid(alpha=0.3)
plt.show()
```

**Interpretation:**

This figure compares average GDP growth rates for treated (high-interest) and control countries before and after the 2017 policy intervention. In the pre-policy period, treated countries exhibit lower average GDP growth than control countries. Following the policy, GDP growth increases for both groups, with a slightly larger increase observed for treated countries. However, the before–after comparison is descriptive and does not account for confounding factors; formal Difference-in-Differences estimates indicate that the post-policy differential is not statistically significant.

```python
event_df = df.copy()
event_df["time_to_treatment"] = event_df["year"] - 2017

event_plot = event_df.groupby("time_to_treatment")["GDP Growth Rate (%)"].mean().reset_index()

plt.figure(figsize=(10,6))
sns.lineplot(data=event_plot, x="time_to_treatment", y="GDP Growth Rate (%)", marker="o")
plt.axvline(0, color="red", linestyle="--")
plt.title("Event Study Plot: GDP Growth Around Policy Change")
plt.xlabel("Years Since Policy")
plt.grid(True)
plt.show()
```

**Interpretation:**

This event-study plot shows average GDP growth relative to the timing of the policy intervention, normalized to event time zero. The vertical dashed line denotes the year of the policy change. GDP growth in the pre-treatment period exhibits no systematic upward or downward trend, providing visual support for the parallel trends assumption. Post-treatment, GDP growth increases modestly before stabilizing, though subsequent regression-based estimates indicate that these post-policy changes are not statistically significant.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

plt.figure(figsize=(8,6))
plt.scatter(D_res, Y_res, alpha=0.5)
plt.title("DML Residual Plot: D_res vs Y_res")
plt.xlabel("Residualized Interest Rate (D_res)")
plt.ylabel("Residualized GDP Growth (Y_res)")

# regression line
lr = LinearRegression().fit(D_res.reshape(-1,1), Y_res)
x_vals = np.linspace(D_res.min(), D_res.max(), 100)
plt.plot(x_vals, lr.predict(x_vals.reshape(-1,1)), color="red", linewidth=2, label="DML Slope")

plt.legend()
plt.grid(True)
plt.show()

print("DML Estimated Causal Effect:", lr.coef_[0])

```

**Interpretation:**

This figure plots residualized GDP growth against residualized interest rates from the Double Machine Learning (DML) procedure. The residuals represent variation unexplained by high-dimensional controls. The fitted line corresponds to the DML estimate of the causal effect. The near-flat slope indicates a weak relationship between interest rates and GDP growth, consistent with the small and statistically insignificant DML coefficient.
